### 모델 학습

```python
from ultralytics import YOLO

# 1. 모델 로드
model = YOLO("yolov8n.pt")

# 2. 학습 시작
model.train(
    data="FireSmokeDetection-4/data.yaml",
    epochs=100,
    patience=10,
    imgsz=640,
    batch=16, # 메모리에 따라 조절 (CPU=4~8, GPU=16~32)
    project="runs", # 결과 저장 경로
    name="firesmoke_train", # 실험 이름
    resume=False # 중단된 학습 이어서 할 건지 여부
)
```
